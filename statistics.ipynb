# probability

1. Basic Probability
Definition: Probability measures the likelihood of an event occurring.

P(A)= Total number of possible outcomes / Number of favorable outcomes

Example: Rolling a Die

What is the probability of rolling a 4?


```python
# What is the probability of rolling a 4?

total_outcomes = 6

favourable_outcomes = 1

probability = favourable_outcomes / total_outcomes

print(f'probability of rolling 4 is : {probability:.2f}')
```

2. Addition Rule of Probability
Formula:
For two events 
A and 𝐵

P(A∪B)=P(A)+P(B)−P(A∩B)
Example: Drawing a card that is either a King or a Heart

```python
# Total cards in a deck
total_cards = 52

# Cards that are Kings
kings = 4

# Cards that are Hearts
hearts = 13

# Overlap (King of Hearts)
overlap = 1

# Probability of King or Heart
P_king_or_heart = (kings / total_cards) + (hearts / total_cards) - (overlap / total_cards)
print(f"Probability of drawing a King or a Heart: {P_king_or_heart:.2f}")
```

3. Multiplication Rule of Probability

P(A∩B)=P(A)⋅P(B)

Example: Tossing Two Coins

What is the probability of getting heads on both coins?

```python
# Probability of heads on one coin
P_heads = 0.5

# Probability of heads on both coins
P_both_heads = P_heads * P_heads
print(f"Probability of getting heads on both coins: {P_both_heads:.2f}")
```

4. Conditional Probability

P(A∣B)= P(A∩B)/P(B)


A bag has 3 red balls and 2 blue balls. What is the probability of picking a red ball given the first ball was red?

```python
# Total balls
total_balls = 5

# Probability of first red ball
P_first_red = 3 / total_balls

# Probability of second red ball given first was red
P_second_red_given_first_red = 2 / (total_balls - 1)
print(f"Probability of second red ball given the first was red: {P_second_red_given_first_red:.2f}")
```

5. Bayes' Theorem

Example: Spam Email Detection

Imagine you have an email spam filter. The filter uses specific words to determine whether an email is spam or not.

20% of all emails are spam.

80% of spam emails contain the word "discount."

10% of non-spam emails contain the word "discount."

We want to find the probability that an email is spam given that it contains the word "discount":

P(Spam∣Discount)

P(Spam∣Discount)=  P(Discount∣Spam)⋅P(Spam)/ P(Discount)

P(Discount)=P(Discount∣Spam)⋅P(Spam)+P(Discount∣Not Spam)⋅P(Not Spam)
 


```python
# Given probabilities
P_Spam = 0.2  # Probability of spam
P_Not_Spam = 1 - P_Spam  # Probability of not spam
P_Discount_given_Spam = 0.8  # Probability of "discount" in spam
P_Discount_given_Not_Spam = 0.1  # Probability of "discount" in not spam

# Total probability of "discount"
P_Discount = (P_Discount_given_Spam * P_Spam) + (P_Discount_given_Not_Spam * P_Not_Spam)

# Bayes' Theorem
P_Spam_given_Discount = (P_Discount_given_Spam * P_Spam) / P_Discount

print(f"Probability an email is spam given it contains 'discount': {P_Spam_given_Discount:.4f}")
```

6. Binomial Distribution

What is the probability of getting exactly 3 heads in 5 flips of a fair coin?

Normal Distribution


Example: Heights of People

The average height of people is 170 cm with a standard deviation of 10 cm. What is the probability a person is between 160 and 180 cm?

```python
from scipy.stats import norm

# Parameters
mean = 170
std_dev = 10

# Probability between 160 and 180
P_160_180 = norm.cdf(180, mean, std_dev) - norm.cdf(160, mean, std_dev)
print(f"Probability of height between 160 and 180 cm: {P_160_180:.2f}")
```

# Inferential Statistics Overview

1. Sampling Distributions

Definition:

The sampling distribution is the probability distribution of a statistic (e.g., the sample mean) calculated from multiple random samples of a population.

Real-World Example:

You want to know the average income of people in a city, but surveying everyone is impossible. You take 30 random samples repeatedly and calculate the average income for each sample.

```python
import numpy as np
import matplotlib.pyplot as plt

# Population data (e.g., incomes in $)
population = np.random.normal(50000, 15000, 100000)  # mean=50k, std=15k, size=100k

# Generate sample means
sample_means = []
for _ in range(1000):  # 1000 samples
    sample = np.random.choice(population, size=30)  # sample size = 30
    sample_means.append(np.mean(sample))

# Plot the sampling distribution
plt.hist(sample_means, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title("Sampling Distribution of the Mean")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

2. Point Estimation and Confidence Intervals

Definition:

Point Estimation: A single value (e.g., the sample mean) used as an estimate of a population parameter.

Confidence Interval (CI): A range of values within which the true population parameter likely falls, with a given level of confidence (e.g., 95%).

Real-World Example:

A school wants to estimate the average test score of its students. A random sample of 50 students is selected, and their mean score is calculated. A 95% CI is created to estimate the true average.

Step-by-Step:

Calculate the sample mean and standard deviation.

Use the formula for CI:

CI=  x ± z⋅ n/s

where:

x: Sample mean

z: z-value for confidence level (1.96 for 95%)

s: Sample standard deviation

n: Sample size



```python
import scipy.stats as stats

# Sample data (e.g., test scores)
sample = np.random.normal(75, 10, 50)  # mean=75, std=10, n=50

# Calculate sample mean and standard error
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)
confidence_level = 0.95
z = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Confidence interval
margin_of_error = z * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

3. Hypothesis Testing

Definition:

Hypothesis testing evaluates claims about a population parameter:

Null Hypothesis (𝐻0): The assumption that there's no effect or difference.

Alternative Hypothesis (𝐻𝑎): The claim that there is an effect or difference.

Real-World Example:

A company claims their battery lasts 10 hours. You test this with a sample to see if the claim holds.

Define 𝐻0 : The mean battery life is 10 hours.
Define 𝐻a : The mean battery life is not 10 hours.
Collect sample data and calculate the test statistic (z or t).
Compare the test statistic to the critical value or calculate the p-value.



```python
# Hypothesis: Test if the mean battery life is 10 hours
sample = np.random.normal(10.2, 0.5, 30)  # mean=10.2, std=0.5, n=30
pop_mean = 10  # Claimed mean

# Perform t-test
t_stat, p_value = stats.ttest_1samp(sample, pop_mean)

print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

2. t-Tests

Definition:

A t-test compares the means of two groups to see if they are significantly different.

Example: Comparing Test Scores
You want to check if a new teaching method improves scores.

Group A (old method): [70, 75, 80, 85, 90]
Group B (new method): [75, 80, 85, 90, 95]

```python
from scipy.stats import ttest_ind

# Data
group_a = [70, 75, 80, 85, 90]  # Old method
group_b = [180, 185, 190, 195, 200]  # New method

# Perform t-test
t_stat, p_value = ttest_ind(group_a, group_b)

print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: The new method improves scores.")
else:
    print("Fail to reject the null hypothesis: No improvement.")
```

3. z-Tests

Definition:
A z-test is used when comparing a sample mean to a population mean, and the population standard deviation is known.

Example: Average Battery Life

A company claims their batteries last 10 hours. You test 30 batteries and find an average of 9.8 hours with a standard deviation of 0.5 hours.

```python
from scipy.stats import norm
import numpy as np

# Data
sample_mean = 9.8
population_mean = 10
std_dev = 0.5
n = 30

# Calculate z-statistic
z_stat = (sample_mean - population_mean) / (std_dev / np.sqrt(n))

# Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # Two-tailed test

print(f"Z-Statistic: {z_stat:.2f}")
print(f"P-Value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: The batteries last less than 10 hours.")
else:
    print("Fail to reject the null hypothesis: The batteries meet the claim.")
```

4. Chi-Square Test
Definition:
The Chi-Square test determines if there is a significant association between two categorical variables.

Example: Survey Preferences
You survey 100 people about their preferred drinks: Tea or Coffee, grouped by gender.

Tea	Coffee	Total
Male	30	20	50
Female	10	40	50
Total	40	60	100
Does preference depend on gender?



```python
import numpy as np
from scipy.stats import chi2_contingency

# Observed data
data = np.array([[30, 20],  # Male
                 [10, 40]])  # Female

# Perform Chi-Square Test
chi2, p_value, dof, expected = chi2_contingency(data)

print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-Value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: Preference depends on gender.")
else:
    print("Fail to reject the null hypothesis: Preference does not depend on gender.")
```

5. ANOVA (Analysis of Variance)
Definition:
ANOVA tests whether the means of multiple groups are significantly different.

Real-World Example:
You test if three different fertilizers result in different average crop yields.

```python
# Data: Yields from three fertilizer types
fertilizer_A = [23, 24, 23, 22, 25]
fertilizer_B = [20, 19, 18, 21, 20]
fertilizer_C = [30, 32, 31, 33, 29]

# Perform ANOVA
f_stat, p_value = stats.f_oneway(fertilizer_A, fertilizer_B, fertilizer_C)

print(f"F-Statistic: {f_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis (H0): Means are not all equal.")
else:
    print("Fail to reject the null hypothesis (H0): Means are equal.")
```
